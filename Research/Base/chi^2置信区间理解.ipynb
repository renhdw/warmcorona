```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# 选择不同的自由度
dof_list = [1, 5, 10, 100, 193]

# 设置颜色
colors = ['r', 'g', 'b', 'orange', 'purple']

# x 轴范围（不同 dof 用相同的 x，范围取大一点）
x = np.linspace(0, 250, 1000)

plt.figure(figsize=(10, 6))

# 绘制每条卡方分布曲线及其 99% 截断线
for dof, color in zip(dof_list, colors):
    y = chi2.pdf(x, dof)
    crit = chi2.ppf(0.99, dof)
    plt.plot(x, y, label=f'dof = {dof}', color=color)
    plt.axvline(crit, color=color, linestyle='--', alpha=0.7,
                label=f'99% cutoff (dof={dof}) at {crit:.1f}')

plt.xlabel('χ² value', fontsize=13)
plt.ylabel('Probability density', fontsize=13)
plt.title('Chi-square Distributions with 99% Cutoff for Different dof', fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
