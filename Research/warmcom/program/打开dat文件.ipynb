此代码将测试$te = 1.0,tau = 15.0$的数据平滑过程

```python
import numpy as np
import matplotlib.pyplot as plt
import struct
import os

# 读取 .dat 文件并解析其中的数据
def load_dat_file(file_path):
    """
    读取二进制 .dat 文件并解析其中的数据（假设数据是双精度浮动点数）。
    
    :param file_path: 文件路径
    :return: 解析后的数据（列表）
    """
    # 检查文件是否存在
    if not os.path.exists(file_path):
        print(f"文件 {file_path} 不存在！")
        return None

    # 打开文件并读取内容
    with open(file_path, "rb") as f:
        data = f.read()

    # 计算数据项的数量
    num_items = len(data) // 8  # 假设每个数据项是一个 double 类型，占 8 字节

    # 解析二进制数据，假设文件中的数据是双精度浮动点数
    result = struct.unpack("<" + "d" * num_items, data)

    return result

# 示例：指定要查看的 .dat 文件路径
file_path_en = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\en.dat"
file_path_flux = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\flux.dat"
file_path_de = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\de.dat"

en = load_dat_file(file_path_en)
flux = load_dat_file(file_path_flux)
de = load_dat_file(file_path_de)

file_path_en_10_6_1 = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\10_6_1\calspec\en.dat"
file_path_flux_10_6_1 = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\10_6_1\calspec\flux.dat"
file_path_de_10_6_1 = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\10_6_1\calspec\de.dat"

en_10_6_1 = load_dat_file(file_path_en_10_6_1)
flux_10_6_1 = load_dat_file(file_path_flux_10_6_1)
de_10_6_1 = load_dat_file(file_path_de_10_6_1)

file_path_en_10_7 = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\10_7\calspec\en.dat"
file_path_flux_10_7 = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\10_7\calspec\flux.dat"
file_path_de_10_7 = r"\\wsl.localhost\Ubuntu-22.04\home\hdw\data\monk\plot\warmcorona\test\test_smooth\data\te_1.000_tau_15.000\calspec\10_7\calspec\de.dat"

en_10_7 = load_dat_file(file_path_en_10_7)
flux_10_7 = load_dat_file(file_path_flux_10_7)
de_10_7 = load_dat_file(file_path_de_10_7)

# 读取数据
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter, butter, filtfilt, cheby1, bessel, sosfilt,medfilt
from scipy.ndimage import gaussian_filter1d
from scipy.interpolate import UnivariateSpline
from statsmodels.nonparametric.smoothers_lowess import lowess

# 假设 flux_de_10_6_1 和 en 已经定义好
# 假设原始数据为 flux_de_10_6_1

# 方法 1: Savitzky-Golay 滤波
flux_de_smooth_savgol_10_6_1 = savgol_filter(flux_de_10_6_1, window_length=51, polyorder=3)

# 方法 2: 移动平均
window_size = 51  # 选择窗口大小
flux_de_smooth_ma_10_6_1 = np.convolve(flux_de_10_6_1, np.ones(window_size)/window_size, mode='same')

# 方法 3: 高斯滤波
flux_de_smooth_gaussian_10_6_1 = gaussian_filter1d(flux_de_10_6_1, sigma=5)

# 方法 4: 中值滤波
flux_de_smooth_median_10_6_1 = medfilt(flux_de_10_6_1, kernel_size=51)

# 方法 5: 双指数平滑 (Exponential Moving Average)
alpha = 0.1  # 平滑系数，越小越平滑
flux_de_smooth_exp_10_6_1 = np.zeros_like(flux_de_10_6_1)
flux_de_smooth_exp_10_6_1[0] = flux_de_10_6_1[0]
for i in range(1, len(flux_de_10_6_1)):
    flux_de_smooth_exp_10_6_1[i] = alpha * flux_de_10_6_1[i] + (1 - alpha) * flux_de_smooth_exp_10_6_1[i-1]

# 方法 6: 局部加权回归 (LOWESS)
flux_de_smooth_lowess_10_6_1 = lowess(flux_de_10_6_1, en, frac=0.1)[:, 1]  # LOWESS

# 方法 7: 卷积滤波 (使用高斯核)
flux_de_smooth_convolution_10_6_1 = np.convolve(flux_de_10_6_1, np.exp(-0.5 * (np.arange(-25, 26) / 5)**2), mode='same')

# 方法 8: Spline 拟合
spline = UnivariateSpline(en, flux_de_10_6_1, s=0.5)  # s=0.5 控制平滑程度
flux_de_smooth_spline_10_6_1 = spline(en)

# 方法 9: Chebyshev Filter (高通/低通滤波器)
def chebyshev_filter(data, cutoff=0.1, order=4):
    sos = cheby1(order, 0.5, cutoff, btype='low', output='sos')
    return sosfilt(sos, data)

# 方法 10: Bessel Filter
def bessel_filter(data, cutoff=0.1, order=4):
    sos = bessel(order, cutoff, btype='low', output='sos')
    return sosfilt(sos, data)

# 方法 11: Butterworth Filter
def butter_lowpass(cutoff, fs, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_filter(data, cutoff, fs, order=4):
    b, a = butter_lowpass(cutoff, fs, order)
    return filtfilt(b, a, data)

# 假设 fs 为采样频率，cutoff 为低通滤波器的截止频率
fs = 1  # 假设每个点之间的采样率为1（单位是Hz）
cutoff = 0.1  # 设定低通滤波器的截止频率

# 应用上述方法进行平滑处理
flux_de_smooth_butter_10_6_1 = butter_filter(flux_de_10_6_1, cutoff, fs)
flux_de_smooth_cheby_10_6_1 = chebyshev_filter(flux_de_10_6_1)

# 绘制原始数据和平滑数据
plt.figure(figsize=(10, 6))

# 绘制原始数据
plt.plot(en, flux_de_10_6_1, label="Original Flux_10_6_1", color='b', linestyle='--')

# 绘制平滑后的数据
plt.plot(en, flux_de_smooth_savgol_10_6_1, label="Savitzky-Golay", color='r')
plt.plot(en, flux_de_smooth_ma_10_6_1, label="Moving Average", color='g')
plt.plot(en, flux_de_smooth_gaussian_10_6_1, label="Gaussian Filter", color='c')
plt.plot(en, flux_de_smooth_median_10_6_1, label="Median Filter", color='m')
plt.plot(en, flux_de_smooth_exp_10_6_1, label="Exponential Moving Average", color='y')
plt.plot(en, flux_de_smooth_lowess_10_6_1, label="LOWESS", color='k')
plt.plot(en, flux_de_smooth_convolution_10_6_1, label="Convolution Filter", color='purple')
plt.plot(en, flux_de_smooth_spline_10_6_1, label="Spline Fitting", color='orange')
plt.plot(en, flux_de_smooth_cheby_10_6_1, label="Chebyshev Filter", color='brown')
plt.plot(en, flux_de_smooth_butter_10_6_1, label="Butterworth Filter", color='pink')

# 设置图形参数
plt.xlabel("Energy (keV)")
plt.ylabel("Flux (photons/s)")
plt.title("Energy vs Flux (Smoothed) for flux_de_10_6_1")

# 设置对数刻度
plt.yscale("log")
plt.xscale("log")
plt.grid(True)
plt.legend()

# 显示图形
plt.show()
```
