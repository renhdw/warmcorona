```python
import numpy as np
import matplotlib.pyplot as plt

# 能量范围（keV）
E = np.logspace(-1, 2, 500)  # 从 0.1 keV 到 100 keV

# 不同的 Gamma 值
gamma_values = [1.0, 1.5, 2.0, 2.5, 3.0]

# 绘图
plt.figure(figsize=(8, 6))
for gamma in gamma_values:
    flux = E ** (-gamma)
    plt.plot(E, flux, label=f'Γ = {gamma}')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Energy (keV)', fontsize=14)
plt.ylabel('Flux (arbitrary units)', fontsize=14)
plt.title('Power-law spectra with different photon indices Γ', fontsize=15)
plt.legend()
plt.grid(True, which="both", ls="--", lw=0.5)
plt.tight_layout()
plt.show()
```
