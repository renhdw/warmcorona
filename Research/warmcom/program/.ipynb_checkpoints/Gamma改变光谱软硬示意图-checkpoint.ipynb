```python
import numpy as np
import matplotlib.pyplot as plt

# keV
E = np.logspace(-1, 2, 500)  #  0.1 keV  100 keV

#  Gamma 
gamma_values = [1.0, 1.5, 2.0, 2.5, 3.0]

# 
plt.figure(figsize=(8, 6))
for gamma in gamma_values:
    flux = E ** (-gamma)
    plt.plot(E, flux, label=f' = {gamma}')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Energy (keV)', fontsize=14)
plt.ylabel('Flux (arbitrary units)', fontsize=14)
plt.title('Power-law spectra with different photon indices ', fontsize=15)
plt.legend()
plt.grid(True, which="both", ls="--", lw=0.5)
plt.tight_layout()
plt.show()
```
